# Popularity based recommender system

### Creating a new dataset to contain the total number of votes received by each ISBN number

```python
rat_total=ratings.groupby('ISBN').count()
rat_total.drop(columns=['Book-Rating'],inplace=True)
rat_total.rename(columns={'User-ID':'Total votes'},inplace=True)
rat_total.head()
```

### Creating a new dataset to contain the average rating received by each ISBN number

```python
rat_avg=ratings.groupby('ISBN').mean()
rat_avg.drop(columns=['User-ID'],inplace=True)
rat_avg.rename(columns={'Book-Rating':'Average-Rating'},inplace=True)
rat_avg.head()
```

### Merging total votes dataset with average rating dataset
This is done because an average can also be calculated using just the votes of 2 people. But we want to select only those books which have been highly rated by more than 250 people, so that the book's popularity is actually correct

```python
rat_total_avg = rat_total.merge(rat_avg, on='ISBN')
rat_total_avg.head()
```

```python
rat_books=rat_total_avg.merge(books,on='ISBN')
rat_books.head()
```

```python
pop_df= rat_books[rat_books['Total votes']>=250].sort_values('Average-Rating',ascending=False).head(50)
```

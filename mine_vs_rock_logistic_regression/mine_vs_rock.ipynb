### Imports

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

### Importing the csv file

```python
sonar1 = pd.read_csv("Copy of sonar data.csv",header=None)
```

```python
sonar = sonar1[:]
sonar.head()
```

```python
#checking the number of rows and columns
sonar.shape
```

```python
#checking for null values
sonar.isnull().sum()
```

```python
sonar.describe()
```

```python
sonar[60].value_counts()
```

```python
Y= sonar[60]
X= sonar.drop(columns=60,axis=1)
```

```python
X.head()
```

```python
print(Y[:10])
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,stratify=Y,random_state=10)
```

```python
print(x_train.head(),y_train.head())
```

```python
print(x_test.head(),y_test.head())
```

```python
model= LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_predictions_train = model.predict(x_train)
print(x_predictions_train,y_train)
```

```python
print("Accuracy on training data is:", accuracy_score(x_predictions_train,y_train))
```

```python
x_prediction_test= model.predict(x_test)
print(x_prediction_test[:10],y_test[:10])
```

```python
print("Accuracy with test data :", accuracy_score(x_prediction_test,y_test))
```
